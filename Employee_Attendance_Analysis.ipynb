```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Sample Employee Attendance Dataset
data = {
    "EmployeeID": [101,102,103,104,105,106,107,108,109,110],
    "Name": ["Ali","Sara","Ahmed","Fatima","Usman","Ayesha","Bilal","Zara","Hassan","Noor"],
    "Department": ["HR","IT","IT","Finance","HR","Finance","IT","HR","Finance","IT"],
    "Days_Present": [20,18,22,19,17,21,20,19,22,18],
    "Days_Absent": [2,4,0,3,5,1,2,3,0,4],
    "Late_Arrivals": [1,2,0,1,3,0,2,1,0,2]
}

df = pd.DataFrame(data)
df
```

**Department-wise Attendanc**

```python
dept_attendance = df.groupby("Department")[["Days_Present","Days_Absent"]].sum()
print("Department-wise Attendance:")
print(dept_attendance)
```

**Total Late Arrivals per Department**

```python
late_by_dept = df.groupby("Department")["Late_Arrivals"].sum()
print("\nLate Arrivals by Department:")
print(late_by_dept)
```

**Department-wise Attendance Bar Plot**

```python
dept_attendance.plot(kind="bar", figsize=(6,4))
plt.title("Department-wise Attendance")
plt.ylabel("Days")
plt.show()
```

**Late Arrivals Pie Chart**

```python
late_by_dept.plot(kind="pie", autopct='%1.1f%%', figsize=(6,6))
plt.title("Late Arrivals by Department")
plt.ylabel("")
plt.show()
```

**Insights**

IT department has highest presence and workload

HR department has more late arrivals

Finance department has good attendance but some late arrivals

These insights can help optimize workforce management

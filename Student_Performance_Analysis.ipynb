```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {
    "Student": ["Ali","Sara","Ahmed","Fatima","Usman",
                "Ayesha","Bilal","Zara","Hassan","Noor"],
    "Math": [85,78,92,88,76,95,67,82,90,73],
    "English": [80,85,88,90,70,92,60,75,89,72],
    "Science": [78,82,91,85,74,94,65,80,88,70],
    "Attendance": [90,85,95,92,80,98,70,88,93,75]
}

df = pd.DataFrame(data)
df
```

```python
df.mean(numeric_only=True)
```

```python
df["Total"] = df["Math"] + df["English"] + df["Science"]
df.sort_values(by="Total", ascending=False)
```

```python
df.sort_values(by="Total").head()
```

```python
sns.scatterplot(x="Attendance", y="Total", data=df)
plt.title("Attendance vs Total Marks")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```
